```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.metrics as metrics

df = pd.read_csv("2025jul18.csv")

print(df)
```

```python
data = df.to_numpy()
```

```python
plt.scatter(df['Current'], df['Voltage(1)'])
plt.scatter(df['Current'], df['Voltage(2)'])
plt.scatter(df['Current'], df['Voltage(3)']) 

b1, a1 = np.polyfit(data[:,0], data[:,1], deg=1)
b2, a2 = np.polyfit(data[:,0], data[:,2], deg=1)
b3, a3 = np.polyfit(data[:,0], data[:,3], deg=1)

# Create sequence of 100 numbers from 0 to 100
xseq = np.linspace(data[0,0]-1, data[-1,0]+1, num=100)

# Plot regression line
plt.plot(xseq, a1 + b1 * xseq, color="b", lw=2.5)
plt.plot(xseq, a2 + b2 * xseq, color="r", lw=2.5)
plt.plot(xseq, a3 + b3 * xseq, color="g", lw=2.5)

plt.title('Scatter Plot of Current vs Voltage across different Chokes with different gap sizes')    
plt.xlabel('Current') 
plt.ylabel('Voltage')    
plt.grid()                                 

plt.show() 

print(f'Inductor I, II and III intercept the Y axis at {a1}, {a2}, {a3} respectively')
```

```python
v1_r2 = metrics.r2_score(df['Voltage(1)'], a1 + b1 * xseq[::10])
v2_r2 = metrics.r2_score(df['Voltage(2)'], a2 + b2 * xseq[::10])
v3_r2 = metrics.r2_score(df['Voltage(3)'], a3 + b3 * xseq[::10])
print(f"Voltage(1)'s R2 Score is {v1_r2}")
print(f"Voltage(2)'s R2 Score is {v2_r2}")
print(f"Voltage(3)'s R2 Score is {v3_r2}")
```
